### Applying Models to Prune Graphs

Models were trained and applied to the graphs to prune them using a jacknifing approach, that is, for each graph $i$, a model was trained using all the features except those from graph $i$. The trained model was then applied to graph $i$. The resulting pruned graph was visually examined to determine whether the prunned graph correctly depicts the pulmonary trunk and the right and left pulmonary arteries.

When I say the model was applied to graph $i$, I mean that the model was applied to the features from graph $i$. The decision classificationwas then passed to the editing program to come up with a edited graph.

What about thresholds?

```python
basenames = set(data['basename'])
```

```python
def selectDataAndCreateInteractions(data_):
    data2_ = data_[['exterior2surface', 'pathlength', 'maxdfe', 'depth', 'length2width']]
    data2_['pathlength:depth'] = data_['pathlength'] * data_['depth'] #make interaction term
    data3_ = data2_[['exterior2surface', 'maxdfe', 'depth', 'length2width', 'pathlength:depth']]

    return data3_
```

```python
def fitModel(classifier, data, targets):
    #predictors = np.array([data[col] for col in data])
    
    #print "predictors shape",predictors.shape
    #print "targets shape",targets.shape
    predictors = data.as_matrix()
    classifier.fit(predictors,targets)
    return classifier
```

```python
data = pd.read_csv(file_, sep="\t",na_values=['None'])
data = data.fillna(data.mean())
prunningResults = {}
for b in list(basenames):
    print b+" ",
    dataTrain = data[data['basename']!=b]
    dataTest = data[data['basename']==b]

    targetTrain = dataTrain[['prune_label']].as_matrix().ravel()
    targetTest = dataTest[['prune_label','edge_label']]

    dataTrain_ = selectDataAndCreateInteractions(dataTrain)

    dataTest_ = selectDataAndCreateInteractions(dataTest)
    fitClassifier = fitModel(RandomForestClassifier(n_estimators=10),dataTrain_,targetTrain)
    #raw_input('continue')
    predictedClasses = fitClassifier.predict(dataTest_)
    prunningResults[b]= zip(targetTest.as_matrix()[:,1],predictedClasses)

    
```

```python
def getGraph(d,i,label="og_rms"):
    try:
        key = (i,label)
        return d[key],key
    except:
        return getGraph(d,i+1,label=label)
    
```

```python
def toBeNamed(sg, key, edges):
    for e in edges:
        print "Current edge",eval(e[0]),e[1]
        if e[1] == 1:
            terminalNode = eval(e[0])[1]
            print "Deleting the following terminal node:",terminalNode
            sg.pruneSpecifiedDegreeOneNode(key,terminalNode)
    sg.remapVoxelsToGraph(key)
```
